```python
import requests 
import pandas as pd
import plotly.express as px
```

```python
start_date = '2020-01-01'
end_date = '2021-05-15'
```

```python
all_prices = pd.DataFrame({'valid_from': [],
                       'valid_to': [],
                       'value_exc_vat': []}) 

url = ('https://api.octopus.energy/v1/products/AGILE-18-02-21/' + 
         'electricity-tariffs/E-1R-AGILE-18-02-21-C/standard-unit-rates/' + 
         f'?period_from={start_date}T00:00Z&period_to={end_date}T02:29Z')

while url is not None: 
    requests.get(url)
    r = requests.get(url)
    output_dict = r.json()
    valid_from = [x['valid_from'] for x in output_dict['results']]
    valid_to =  [x['valid_to'] for x in output_dict['results']]
    value_exc_vat = [x['value_exc_vat'] for x in output_dict['results']]
    prices = pd.DataFrame({'valid_from': valid_from,
                       'valid_to': valid_to,
                       'value_exc_vat': value_exc_vat})
    all_prices = all_prices.append(prices)
    
    # Move to the next page of results
    url = output_dict['next']
 
```

```python
# Save Historic Data to CSV so we don't have to re-run it.
all_prices.to_csv(f'{start_date}_{end_date}_prices.csv')
```
